#PART 6: ANALYZING THE CLINICAL TRIALS DATASET

```python
clinical_trials['clinical_trial_exp'] = True 
#This code here confirms the doctor's identify whether he/she has clinical trial experience.
```

```python
clinical_trials.head(10) # here is the clinical trials dataframe
```

```python
clinical_trials = clinical_trials.rename({"name_x":"facility_nm"}, axis = 1)
clinical_trials = clinical_trials.rename({"name_y":"name"},axis = 1)
clinical_trials = clinical_trials.rename({"cty":"city"}, axis = 1)
clinical_trials = clinical_trials.rename({"st":"state"}, axis = 1)
clinical_trials.head()
#Just cleaning up the names of the columns
```

```python
clinical_trials[['name','Cred']]= clinical_trials['name'].str.split(',', n=1, expand = True)
clinical_trials.head(10) #here we are splitting name as their following cred/credentials
```

```python
clinical_trials['array'] = clinical_trials['name'].str.split(' ')
```

```python
clinical_trials.head()
```

```python
clinical_trials['first_nm']= clinical_trials['array'].apply(lambda x: x[0]) # array split by there first name column
clinical_trials['last_nm']= clinical_trials['array'].apply(lambda x: x[-1]) # array split by there last name column  
```

```python
clinical_trials = clinical_trials.rename({"frst_nm":"first_nm"}, axis = 1)
clinical_trials = clinical_trials.rename({"lst_nm":"last_nm"},axis = 1)
clinical_trials = clinical_trials.rename({"CRED":"Cred"},axis = 1)
clinical_trials.head()
# renaming clinical trials columns properly 
```

```python
clinical_trials['first_nm']= clinical_trials['first_nm'].str.upper()
clinical_trials['last_nm']= clinical_trials['last_nm'].str.upper()
# changing the first and last names of the doctors from lower case to upper case so it all merges and navigates its match correctly
```

```python
clinical_trials.drop(['array'],axis = 1) # here i dropped the array column, becuase it wasnt useful since i split them by there first and last name 
```
